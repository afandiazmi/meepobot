## 1.导入必要的库文件

```python
from LOBOROBOT import LOBOROBOT  # 载入机器人库
import pylirc as lirc
import time
```

## 2. 定义机器人对象

```python
clbrobot = LOBOROBOT() # 实例化机器人对象
makerobo_blocking = 0 # 判断值
```

## 3.初始化函数

```python
# GPIO初始化设置
def makerobo_setup():
	lirc.init("pylirc", "/etc/lirc/conf", makerobo_blocking) # 载入配置参数
```

## 4 .红外遥控控制函数

```python
def robot_control(config):
    global color
    # 按键第一行第一个
    if config == 'KEY_CHANNELDOWN':
        clbrobot.forward_Left(50,3) # 机器人前左斜
        clbrobot.t_stop(1)          # 停止
        print ('Robot front left oblique')
    if config == 'KEY_CHANNEL':     # 按键第一行第二个
        clbrobot.t_up(50,3)         # 机器人前进
        clbrobot.t_stop(1)          # 停止
        print ('Robot advance')
    if config == 'KEY_CHANNELUP':     # 按键第一行第三个
        clbrobot.forward_Right(50,3)  # 机器人前右斜
        clbrobot.t_stop(1)          # 停止
        print ('Robot front right tilt')
    if config == 'KEY_PREVIOUS':    # 第二行第一个
        clbrobot.moveLeft(50,3)
        clbrobot.t_stop(1)          # 停止
        print ('Robot shift left')
    if config == 'KEY_NEXT':       # 第二行第二个
        clbrobot.t_stop(1)         # 机器人停止
        print ('Robot stop')
    if config == 'KEY_PLAYPAUSE':  # 第二行第三个
        clbrobot.moveRight(50,3)   # 机器人右移
        clbrobot.t_stop(1)          # 停止
        print ('Robot right shift')
    if config == 'KEY_VOLUMEDOWN': # 第三行第一个
        clbrobot.backward_Left(50,3)   # 机器人后左斜
        clbrobot.t_stop(1)             # 停止
        print ('Robot back left oblique')
    if config == 'KEY_VOLUMEUP':  # 第三行第二个
        clbrobot.t_down(50,3)    # 机器人后退
        clbrobot.t_stop(1)          # 停止
        print ('Robot retreat')
    if config == 'KEY_EQUAL':    # 第三行第三个
        clbrobot.backward_Right(50,3) # 机器人后右斜
        clbrobot.t_stop(1)          # 停止
        print ('Robot back right oblique')
```

## 7.循环函数

```python
# 循环函数
def makerobo_loop():
	while True:
		s = lirc.nextcode(1)    # 获取红外遥控器码值		
		while(s):
			for (code) in s:
				print ("Command: ", code["config"]) # 调试信息，可以具体知道按下了哪个按键
				robot_control(code["config"])  # 调用控制RGB函数
			if(not makerobo_blocking):       # 读取到值
				s = lirc.nextcode(1)       # 再一次获取红外遥控器码值	
			else:
				s = []
```
